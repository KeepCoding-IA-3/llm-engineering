```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import time
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
embedder = SentenceTransformer("all-MiniLM-L6-v2")
documents = [
    "Retrieval-Augmented Generation (RAG) combines language models with external document retrieval to improve factual accuracy.",
    "Context-Aware Generation (CAG) relies on supplying relevant background info in the prompt.",
    "FAISS enables fast similarity search across large text embeddings.",
    "Retrieval helps language models by grounding answers in actual data, reducing hallucination.",
    "Static context in prompts allows models to focus on specific themes but lacks flexibility."
]
doc_embeddings = embedder.encode(documents, convert_to_numpy=True)
index = faiss.IndexFlatL2(doc_embeddings.shape[1])
index.add(doc_embeddings)
```

```python
query = "How does retrieval help improve generation?"
query_embedding = embedder.encode([query])
```

## RAG

```python
rag_start = time.time()
_, indices = index.search(query_embedding, k=2)
retrieved_docs = "\n".join([documents[i] for i in indices[0]])
rag_prompt = f"""You are a helpful assistant.
Use the following context to answer the question concisely and clearly.
Context:
{retrieved_docs}
Question: {query}
Answer:"""
rag_inputs = tokenizer(rag_prompt, return_tensors="pt", truncation=True)
rag_output = model.generate(**rag_inputs, max_new_tokens=100)
rag_response = tokenizer.decode(rag_output[0], skip_special_tokens=True)
rag_end = time.time()
rag_time = rag_end - rag_start
rag_prompt_tokens = len(tokenizer.tokenize(rag_prompt))
rag_output_tokens = len(tokenizer.tokenize(rag_response))
```

## CAG

```python
cag_start = time.time()

cag_context = (
    "Retrieval gives language models access to external knowledge, helping them reduce hallucination "
    "and improve factual accuracy during response generation."
)
cag_prompt = f"""You are a helpful assistant.
Use the provided context to answer the question.
Context:
{cag_context}
Question: {query}
Answer:"""
cag_inputs = tokenizer(cag_prompt, return_tensors="pt", truncation=True)
cag_output = model.generate(**cag_inputs, max_new_tokens=100)
cag_response = tokenizer.decode(cag_output[0], skip_special_tokens=True)
cag_end = time.time()
cag_time = cag_end - cag_start
cag_prompt_tokens = len(tokenizer.tokenize(cag_prompt))
cag_output_tokens = len(tokenizer.tokenize(cag_response))
```

## Conclusiones

```python
print("QUERY:", query)
print("\n=== COMPARISON ===\n")

print("RAG")
print("Prompt Tokens:", rag_prompt_tokens)
print("Generated Tokens:", rag_output_tokens)
print("Time Taken:", round(rag_time, 3), "s")
print("Output:", rag_response)
print("\nCAG")
print("Prompt Tokens:", cag_prompt_tokens)
print("Generated Tokens:", cag_output_tokens)
print("Time Taken:", round(cag_time, 3), "s")
print("Output:", cag_response)
```

## Diferencia entre RAG y CAGar

### RAG (Retrieval-Augmented Generation):
Recupera dinámicamente documentos relevantes desde un almacén vectorial durante la generación. Permite respuestas más precisas, basadas en datos reales. Es ideal cuando se necesita precisión o información externa actualizada.

### CAG (Context-Aware Generation):
Inyecta un contexto estático seleccionado previamente directamente en el prompt. Es más rápido y eficiente, pero depende totalmente de la calidad del contexto predefinido.

El CAG (Context-Aware Generation) es importante porque aporta una forma eficiente, controlada y predecible de mejorar las respuestas de un modelo de lenguaje.
